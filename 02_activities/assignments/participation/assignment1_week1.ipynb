```python
#import important libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# create data set

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint (0,100,50)
y2 = np.random.randint (0,100,50)


```

```python
# plot both y-axis on same line graph

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1)
ax.plot(x,y2)
```

```python
# adding legend to graph

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1, label = 'Person 1')
ax.plot(x,y2, label = 'Person 2')
ax.legend(loc = 'lower right')
```

```python
# modifying legend of graph


fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1, label = 'Person 1')
ax.plot(x,y2, label = 'Person 2')
ax.legend(loc = 'lower right', 
    frameon = True,
    fontsize = 12,
    ncol = 2,
    shadow = True)
```

```python
# moving location of legend 

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1, label = 'Person 1')
ax.plot(x,y2, label = 'Person 2')
ax.legend(loc = 'upper right',   # specify where legend box stays
          bbox_to_anchor = (1,1)) # finetune location of legend box - 1 is 100% of figure and 0.5 is 50% of our figure
```

```python
# annotating graph with text

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = 'Person 1')
ax.scatter(x,y2, label = 'Person 2')
ax.legend(loc = 'lower right')
ax.text(10,95, 'This value is important!',   
        ha = 'center',
        color = 'red',
        size = 20)         # specify coordinate on axis where text should be (relevant to data point), alignment, color and size
```

```python
# postioning text annotations

fig, ax = plt.subplots()
ax.axis([0,10,0,10])

ax.text(1,5, ". Data: (1,5)", transform = ax.transData)  # transform tells us the coordinates are referencing the data points
ax.text(0.5,0.1,". Axes: (0.5,0.1)", transform = ax.transAxes) # transform tells us the coordinates are referencing the axes points
ax.text(0.2,0.2,".Figure: (0.2,0.2)", transform = fig.transFigure) # transform tells us the coordinates are referencing the overall figure
```

```python
# annotating graph with arrow

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1,label ='Person 1')
ax.scatter(x,y2,label ='Person 2')
ax.legend(loc= 'lower right')


ax.annotate('This is important', xy =(10,95),
            xytext= (20,94),
            arrowprops= dict(facecolor= 'black', shrink=0.05))  # xy is referencing data point we are calling out and xytext references text location
```

```python
# annotating graph with arrow + changing arrow style

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1,label ='Person 1')
ax.scatter(x,y2,label ='Person 2')
ax.legend(loc= 'lower right')


ax.annotate('This is important', xy =(10,95),
            xytext= (20,94),
            arrowprops= dict(facecolor= 'yellow', arrowstyle='wedge'))
```

```python
# removing tick marks and labels

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1,label ='Person 1')
ax.scatter(x,y2,label ='Person 2')

ax.yaxis.set_major_locator(plt.NullLocator()) #locator to remove ticks and labels
ax.yaxis.set_major_formatter(plt.NullFormatter()) #formatter to remove only labels
```

```python
#editing appearance of tick marks

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1,label ='Person 1')
ax.scatter(x,y2,label ='Person 2')
ax.legend(loc= 'lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3)) #tells matplotlib we only want 3 ticks on our x axis
```

```python
#setting tick marks to appear at specified intervals

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1,label ='Person 1')
ax.scatter(x,y2,label ='Person 2')
ax.legend(loc= 'lower right')
#ax.xaxis.set_major_locator(plt.MaxNLocator(3)) 
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # setting tick makr intervals to be every 5 units
```

```python
#rotating axis labels

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1,label ='Person 1')
ax.scatter(x,y2,label ='Person 2')
ax.legend(loc= 'lower right')

plt.xticks(rotation= 45, ha= 'right',
           color= 'pink', 
           size= 20)
```

```python
#modify axis titles and how they look like
font1 = {'family':'serif', 'color':'indigo'}

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y1,label ='Person 1')
ax.scatter(x,y2,label ='Person 2')
ax.legend(loc= 'lower right')

plt.xticks(rotation= 45, 
           ha= 'right',
           color= 'pink', 
           size= 20)

plt.xlabel('Shiny New Label!', fontsize = 18, fontdict = font1)
```

```python
# importing new packages and creating new data set

import PIL
import requests

np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(['Luffy', 'Zoro','Nami','Usopp', 'Sanji'])
y2 = np.array([110,180,240,99,220])
```

```python
# making figure container with subplots

fig, (ax1, ax2) = plt.subplots(ncols=2,
      nrows=1,
      figsize=(7,3)) #defining spacial layout of figure, 2 columns, 1 row

ax1.scatter(x1,y1)  # defining graph style and reference data points for each subplot
ax2.bar(x2,y2)
```

```python
#customizing subplots from above

fig, (ax1, ax2) = plt.subplots(ncols=2,
      nrows=1,
      figsize=(7,3)) #defining spacial layout of figure, 2 columns, 1 row

ax1.scatter(x1,y1,color = 'purple')  # defining graph style and reference data points for each subplot
ax2.bar(x2,y2,color = 'pink')

plt.style.use('dark_background')

```

```python
# making subplots without a grid arrangement

fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                    ['ax2','ax3']],
                                    figsize=(7,4)) #specifying placement of subplot and figure size

someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1) #specifying data to be referenced for each plot
```

```python
# tweaking layout of 3 plot mosaic to illustrate possible problems with titles

fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                    ['ax2','ax3']],
                                    figsize=(7,4))

someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)

someaxes['ax1'].set_xlabel('A Big Label', fontsize= 18)
someaxes['ax2'].set_xlabel('Another Label', fontsize= 18)
someaxes['ax3'].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize= 18)
```

```python
# fixing layout issue from above ex tight/constrained layouts

fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                    ['ax2','ax3']],
                                    figsize=(7,4),
                                    layout='constrained')

someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)

someaxes['ax1'].set_xlabel('A Big Label', fontsize= 18)
someaxes['ax2'].set_xlabel('Another Label', fontsize= 18)
someaxes['ax3'].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize= 18)
```
