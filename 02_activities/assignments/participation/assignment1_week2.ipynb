```python
# import necessary libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
from PIL import Image
import requests
from io import BytesIO
```

```python
# plotting to graphs in one figure

x = np.array(['Luffy','Zoro','Nami','Usopp','Sanji'])
y1 = np.array ([110,180,240,99,220])
y2 = np.array ([170,100,99,120,50])

fig, ax = plt.subplots(figsize=(7,3))

ax.bar (x,y1, color = 'indigo')
ax.plot (x, y2, color = 'red')
```

```python
# add error bars to show standard deviation of data

y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize = (7,3))
ax.plot(x,y2, color= 'red')

ax.errorbar(x,y2,
            yerr = y2_sd,
            fmt = 'none') # yerr calculates standard dev of y2
```

```python
# adding modification to error bars ex color, width, interval etc

y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize = (7,3))
ax.plot(x,y2, color= 'red')

ax.errorbar(x,y2,
            yerr = y2_sd,
            fmt = 'none',
            ecolor= 'indigo',
            elinewidth= 4,
            capsize= 6,
            capthick= 4,
            errorevery= 2)
```

```python
# Grabbing image from the internet

response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
# adding an image to a plot

fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x,y2, color= 'red')

ax_image = fig.add_axes([0.1,0.11,0.15,0.35]) # setting location for container- x coordinate, y coordinate, then width and height of the axis container for image
```

```python
# modifying visual of image container

fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x,y2, color= 'red')

ax_image = fig.add_axes([0.1,0.11,0.15,0.35]) # setting location for container- x coordinate, y coordinate, 
                                              # then width and height of the axis container for image
                                            
ax_image.imshow(image) # placing imported image from internet into the plot
ax_image.axis('off')   # turn off border/ticks for image container                                         
```

```python
# import necessary libraries 

import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# load sample data from seaborn

tips = sns.load_dataset('tips')
print(tips)
```

```python
# making line plot using tips data from seaborn package

sns.lineplot(data = tips,
                x = 'total_bill',
                 y = 'tip')
```

```python
# setting grid style to seaborn plot

sns.set_style('darkgrid')

sns.lineplot(data = tips,
                x = 'total_bill',
                 y = 'tip')
```

```python
# setting title and labels to seaborn graph

tipgraph = sns.lineplot(data= tips,
                        x = 'total_bill',
                        y = 'tip')

tipgraph.set(title= 'Tips vs. Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip Amount ($)')
```

```python
# modify aesthetic elements to plot

fig = plt.subplots(figsize =(10,3))

tipgraph = sns.lineplot(data = tips,
                        x= 'total_bill',
                        y= 'tip',
                        color= 'hotpink',
                        linestyle= '--',
                        linewidth = 3,
                        marker= 'o',
                        markerfacecolor= 'indigo')

tipgraph.set(title= 'Tips vs Total Bill',
             xlabel= 'Total Bill ($)',
             ylabel= 'Tip Amount ($)')
```

```python
# add multiple variables to a plot

tipgraph = sns.scatterplot(data = tips,
                        x= 'total_bill',
                        y= 'tip',
                        style= 'time',  
                        hue= 'day',
                        palette= ['purple', 'hotpink','deepskyblue','yellowgreen'])

# style = time means change the style for every diff time data point and hue means change the colour for every day data point

tipgraph.set(title= 'Tips vs Total Bill',
             xlabel= 'Total Bill ($)',
             ylabel= 'Tip Amount ($)')
```

```python
# using pairplot to explore how varibales influence each other

sns.pairplot(data=tips,
            hue= 'day') #relationship between day variable and tip trend
```

```python
# replot to explore specific levels within variables

daysplot = sns.relplot(data= tips,
                      x= 'total_bill',
                      y= 'tip',
                      hue= 'sex',
                      col= 'day',
                      kind= 'scatter',
                      col_wrap= 2) # reviewing total bill vs tip, on each day and by gender of each customer

daysplot  
```

```python
#import libraries and load data

import plotly.graph_objects as go

x1= np.array(['Luffy', 'Zoro','Nami','Usopp','Sanji'])
y1= np.array([110,180,240,99,220])
```

```python
# making plot with plotly and updating layout

graph= go.Figure()

graph.add_trace(go.Bar(x= x1, y= y1))

graph.update_layout(title = 'Pirate Scores',
                    xaxis_title= 'Pirates',
                    yaxis_title= 'Score')

# creates interactive data point swhen you hover over graphs and also provides other options in top right corner for interacting with the graph
```

```python
# making interactive scatter plot with plotly and updating layout

graph= go.Figure()

graph.add_trace(go.Scatter(
                x= x1,
                y= y1,
                mode= 'markers',
                marker= dict(
                    size= 15,
                    color= 'hotpink',
                    opacity= 1,
                    line= dict(width= 5,
                               color= 'purple')
                )
))

graph.update_layout(title = 'Interactive Pirate Plot',
                    xaxis_title= 'Pirates',
                    yaxis_title= 'Score',
                    width= 500,
                    height= 500)
```

```python
# making wordclouds and importing package and word file

from wordcloud import WordCloud

df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv")

print(df)
```

```python
text = " ".join(each for each in df.quote) # means for every quote join them together with a space in between

wordcloud= WordCloud(background_color= 'white',
                     colormap= 'inferno').generate(text)

fig, ax = plt.subplots(figsize=(7,3))
ax.imshow(wordcloud,
          interpolation= 'bilinear') # helps image look smoother 
ax.axis('off')
```

```python
# making venn diagram and data set

from matplotlib_venn import venn2, venn2_circles, venn2_unweighted

A= set(['apple','banana','watermelon'])
B= set(['pumpkin','blueberry','apple','key lime'])

diagram = venn2([A,B],
                set_labels = ('Fruits', 'Pies'),
                set_colors = ('blue','red'),
                alpha= 0.5) # alpha refers to transparency of cricles

diagram.get_label_by_id('10').set_text('\n'.join(A-B)) # code specifying postioning of text
diagram.get_label_by_id('11').set_text('\n'.join(A&B))
diagram.get_label_by_id('01').set_text('\n'.join(B-A))
```
